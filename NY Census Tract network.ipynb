```python
import math
import json
import numpy as np
import pandas as pd
import networkx as nx

import matplotlib.pyplot as plt
from IPython.display import Image
%matplotlib inline
```

<strong>Working with nodes </strong>


```python
CTnodes = pd.read_csv('NYCB_Nodes.csv'
)
```

<strong> Working with edges</strong>

```python
CTEdges = pd.read_csv('NYCTEdges.csv')
```

```python
edges = CTEdges[['src_BoroCT2010','nbr_BoroCT2010']].values 
```

```python
print (edges)
```

```python
g = nx.from_edgelist(edges)
```

```python
len(g.nodes()),len(g.edges())
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
nx.draw_networkx(g, ax=ax, node_size=5,
                 font_size=6, alpha=.5,
                 width=.5)
ax.set_axis_off()
```

```python
sg = next(nx.connected_component_subgraphs(g))
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
nx.draw_networkx(sg, ax=ax, with_labels=False,
                 node_size=5, width=.5)
ax.set_axis_off()
```
